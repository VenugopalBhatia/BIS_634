```python
data_raw = pd.read_csv("../Files/incd.csv",header = 4)
data_raw.tail(30)
```

```python
data_raw = data_raw[:53]
```

Replace star and N/A values with np.nan, leaving *** column alone, but the one with the single * were probably suppressed values so have replaced them with NA

```python
data_1 = data_raw.replace("N/A",np.nan,regex = True)
data_1.iloc[:,-4:] = data_1.iloc[:,-4:].replace("*",np.nan)
```

Remove number from state

```python
data_1['State'] = data_1['State'].str.replace("\(\d\)","",regex = True)
```

#### Thus for our data I first removed any extra lines above and below, retained the US(SEER + NCPR), probably could have removed that too, I replaced the "N/A" with np.nan, I replaced the single * in the last 4 columns with np.nan too and finally removed the number alongside the state.
