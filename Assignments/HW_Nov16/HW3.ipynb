```python
from functools import lru_cache
from time import perf_counter
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def fibo(n):
    if(n==1 or n== 2):
        return(1)
    return fibo(n-1) + fibo(n-2)
```

```python
@lru_cache
def fibo_c(n):
    if(n==1 or n== 2):
        return 1
    return fibo_c(n-1) + fibo_c(n-2)
```

```python
def getTime(f,r_max):
    cached_fibo_times = {}
    for i in range(1,r_max):
        t_start = perf_counter()
        val_ = f(i)
        t_stop = perf_counter()
        cached_fibo_times[i] = t_stop - t_start
    return cached_fibo_times
```

```python
fibo_times = getTime(fibo,40)
```

```python
fibo_c_times = getTime(fibo_c,1000)
```

```python
def plot_graph(fibo_times,fibo_c_times,scale = "linear"):
    fig,axs = plt.subplots(2,1, figsize = (10,8),sharey = True)
    fibo_c_vals = sorted(fibo_c_times.items())
    fibo_vals  = sorted(fibo_times.items())
    x,y = zip(*fibo_vals)
    z,c = zip(*fibo_c_vals)
    axs[0].set(xscale = scale,yscale = scale)
    axs[1].set(xscale = scale,yscale = scale)
    sns.scatterplot(x,y,ax = axs[0])
    sns.scatterplot(z,c,ax = axs[1])
  
    
    
```

#### Linear Scale

```python
plot_graph(fibo_times,fibo_c_times)
```

#### Log Log graphs

```python
plot_graph(fibo_times,fibo_c_times,scale = "log")
```

#### As we saw in class the time complexity for a recursive fibonacci sequence will be of the order of 1.6^n or approximating the golden ratio. This can be attributed to the fact that for n we calculate n-1 and n-2 terms and for each of those we calculate their n-1 and n-2 terms. Thus we end up calculating the same thing repeatedly and we may express time taken at N as T(N) = T(N-1) + T(N-2) ~ 2T(N-1). Thus on solving further we can see a time complexity that is of the order of x^n. 

#### Using lru_cache we can cache values and thus we only really need to calculate them once after which point the method looks up the value in the stored cache. This results in almost negilgible time for even very large values of n. In comparison the regular fibonacci function takes a lot longer to calculate for n = 50 than it does for n = 35.

#### I plotted a linear scale scatter plot and a scatter plot on a log-log scale. The x axis varies since I ran the cached version for n = 1000, and it still took time close to 0. The regular method quickly starts to grow beyond 30. Even on a log-log graph if time complexity were polynomial say n^2 we'd see a graph that's a straight line with a slope but the fact that it still looks exponential on a log-log graph sort of validates the time complexity of ~1.6^n
